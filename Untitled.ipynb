```python
import os
import glob
import cv2
import numpy as np
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data_dir = 'Test/Test'  # Replace with the actual path

image_paths = glob.glob(os.path.join(data_dir, '**', '*.jpg'), recursive=True)
```

```python
images = []
labels = []
for image_path in image_paths:
    try:
        image = cv2.imread(image_path)  # Read image using OpenCV
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) #Convert to RGB
        image = cv2.resize(image, (224, 224)) #Resize images if required
        images.append(image)
        
         # Extract label from folder structure (assuming subfolders are class names)
        label = os.path.basename(os.path.dirname(image_path))
        labels.append(label)

    except Exception as e:
        print(f"Error reading image {image_path}: {e}")

images = np.array(images)
labels = np.array(labels)
```

```python
# Check if labels were extracted correctly
unique_labels = np.unique(labels)
print(f"Unique labels found: {unique_labels}")

if len(unique_labels) != 2:
    print("Error: Incorrect number of labels detected. Ensure correct folder structure.")
    exit()
```

```python
# The version you asked about (encoding unique labels):
le_unique = LabelEncoder()
encoded_unique_labels = le_unique.fit_transform(unique_labels)
print("Encoded Unique Labels:", encoded_unique_labels)

# The correct way to encode ALL labels for a classification task:
le_all = LabelEncoder()
encoded_all_labels = le_all.fit_transform(labels)
print("Encoded All Labels:", encoded_all_labels)

print(f"Loaded {len(images)} images")
print(f"Loaded {len(encoded_all_labels)} labels")

#Inverse transform
original_all_labels= le_all.inverse_transform(encoded_all_labels)
print("Original All Labels:", original_all_labels)
```

```python
import random

def create_pairs_efficient(fire_images, non_fire_images, num_pairs):
    pairs = []
    labels = []
    created_pairs = set() #Set to store created pairs efficiently

    def add_pair(img1, img2, label):
        sorted_pair = tuple(sorted((img1, img2)))
        if sorted_pair not in created_pairs:
            pairs.append((img1, img2))
            labels.append(label)
            created_pairs.add(sorted_pair)
            return True
        return False

    # Positive pairs
    for _ in range(num_pairs // 2):
        while True:
            img1 = random.choice(fire_images)
            img2 = random.choice(fire_images)
            if add_pair(img1, img2, 1):
                break

        while True:
            img1 = random.choice(non_fire_images)
            img2 = random.choice(non_fire_images)
            if add_pair(img1, img2, 1):
                break

    # Negative pairs
    for _ in range(num_pairs):
        while True:
            img1 = random.choice(fire_images)
            img2 = random.choice(non_fire_images)
            if add_pair(img1, img2, 0):
                break

    return pairs, labels
```

```python
fire_image_dir = "Test/Test/Fire"
non_fire_image_dir = "Test/Test/No_Fire"
fire_images = glob.glob(os.path.join(fire_image_dir, "*.jpg")) + glob.glob(os.path.join(fire_image_dir, "*.png"))
non_fire_images = glob.glob(os.path.join(non_fire_image_dir, "*.jpg")) + glob.glob(os.path.join(non_fire_image_dir, "*.png"))
pairs, labels = create_pairs_efficient(fire_images, non_fire_images, 1000) #Creates 1000 negative pairs and 500 of each positive pairs

# Now 'pairs' contains tuples of image paths, and 'labels' contains corresponding labels
```

# Creating siamese network

```python
import torch
import torch.nn as nn
import torchvision.models as models
import torch.nn.functional as F
```

```python
class SiameseNetwork(nn.Module):
    def __init__(self, embedding_size=128):  # Embedding size is a hyperparameter
        super(SiameseNetwork, self).__init__()

        # Load pre-trained ResNet50 (excluding the final fully connected layer)
        resnet = models.resnet50(pretrained=True)
        self.base_network = nn.Sequential(*list(resnet.children())[:-1]) #Removes last FC layer

        # Add a fully connected layer to create the embedding
        self.embedding = nn.Sequential(
            nn.Linear(resnet.fc.in_features, embedding_size),
            nn.ReLU()
        )

    def forward_once(self, x):
        # Forward pass of one subnetwork
        output = self.base_network(x)
        output = output.view(output.size(0), -1) #Flatten the output
        output = self.embedding(output)
        return output

    def forward(self, input1, input2):
        # Forward pass of the Siamese network
        output1 = self.forward_once(input1)
        output2 = self.forward_once(input2)
        return output1, output2
```
